```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#For dropping unamed column
#df =df.drop("Unnamed: 0", axis = 1)

print(df.head())


```

```python
#Data Transformation
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","8-10")

find_df= df[df["WklyStudyHours"]=="8-10"]
print(find_df)
```

```python
#Checking Gender Distribution
plt.figure(figsize=(4,4))
ax= sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution")
#For labels
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#Group by

gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore": 'mean',
                                   "WritingScore" :'mean'})
print(gb)

                                   
```

```python
#Heatmap for analysing the results above 
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True) #annot is for the marks label
plt.title("Degree Impact on the Student Results")
plt.show()


#Results show that higher degree gets more children marks..
```

```python
#Now analyze if martial status affects the student results
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean'
                                            ,"WritingScore":'mean'})

print(gb1)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb1,annot=True)
plt.title("Students impact on the Marital Staus of their Parents")

plt.show()

#Not much difference between them
```

```python
gb2 = df.groupby("IsFirstChild").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb2)

```

```python
sns.heatmap(gb2, annot=True)
plt.figure(figsize=(4,4))
plt.show()
```

```python
#WE use Boxplot for checking Outliers

sns.boxplot(data=df, x= "MathScore")
```

```python
#Checking unique Values in Ethnic group
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df['EthnicGroup']=="group A")].count() 
groupB = df.loc[(df['EthnicGroup']=="group B")].count() 
groupC = df.loc[(df['EthnicGroup']=="group C")].count() 
groupD = df.loc[(df['EthnicGroup']=="group D")].count() 
groupE = df.loc[(df['EthnicGroup']=="group E")].count() 

labelling = ["Group A","Group B","Group C","Group D","Group E"]

mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"],groupC["EthnicGroup"],
         groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels =labelling, autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
          
                 
```

```python
print(df["IsFirstChild"].unique())
```

```python
contingency_table = pd.crosstab(df['IsFirstChild'], df['PracticeSport'], normalize='index')
print(contingency_table)
```

```python
contingency_table.plot(kind='bar', stacked=True)
plt.title('PracticeSport Distribution for First Child')
plt.xlabel('Is First Child')
plt.ylabel('Proportion')
plt.legend(title='PracticeSport')
plt.show()
```

```python
#Same Representation using seaborn
sns.countplot(data=df, x='IsFirstChild', hue='PracticeSport')
plt.title('PracticeSport Distribution for First Child')
plt.xlabel('Is First Child')
plt.ylabel('Count')
plt.show()
```
